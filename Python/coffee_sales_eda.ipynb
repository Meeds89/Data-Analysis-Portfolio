# Coffee Sales EDA

### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing and reading Dataframe

```python
df = pd.read_excel('coffee_sales_cleaned.xlsx')
df.head()
```

### Searching for Best-Selling Coffees

```python
df.groupby(['coffee_name'])['coffee_name'].count().sort_values(ascending=False)
```

```python
# Visualizing
plt.bar(df.groupby(['coffee_name'])['coffee_name'].count().index, df.groupby(['coffee_name'])['money'].count())
plt.xticks(rotation=90)
plt.title('Best-Selling Coffees')
plt.ylabel('Sales')
plt.xlabel('Coffees')
plt.show()
```

### Most Used Payment Method

```python
df.groupby(['cash_type'])['coffee_name'].count()
```

```python
# Visualizing
plt.bar(df.groupby(['cash_type'])['cash_type'].count().index, df.groupby(['cash_type'])['cash_type'].count())
plt.title('Most Used Payment Method')
plt.ylabel('Sales')
plt.xlabel('Payment Method')
plt.show()
```

### Monthly Revenue

```python
df.groupby(df['date'].dt.month)['money'].sum()
```
